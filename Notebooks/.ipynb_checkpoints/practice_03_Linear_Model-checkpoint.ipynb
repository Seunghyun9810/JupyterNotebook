# 1. Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_wave(n_samples=60)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
type(lr.coef_)
```

```python
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

- 특성(가중치)가 하나일 때는 모델이 지나치게 단순.
> 과소적합 발생1

```python
# boston real estate price
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

lr = LinearRegression().fit(X_train, y_train)

print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

# 2. Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)    # alpha=1.0(default)

print(ridge.score(X_train, y_train))
print(ridge.score(X_test, y_test))
```

```python
# ridge accuracy when alpha=0.1
ridge_02 = Ridge(alpha=0.1).fit(X_train, y_train)

print(ridge_02.score(X_train, y_train))
print(ridge_02.score(X_test, y_test))
```

```python
# learning curve
mglearn.plots.plot_ridge_n_samples()
```

* 특이한 점
1. 전체적으로 선형 회귀가 규제가 있는 리지 회귀보다 훈련 성능이 높다.
2. 그러나 테스트 데이터에서는 리지 회귀의 성능이 훨씬 높다.
3. 데이터가 많아질수록 두 모델의 성능이 좋아지고, 마지막에는 선형 회귀가 리지 회귀를 따라잡는다.

> verdict: 데이터를 충분히 많이 주면 규제 항(alpha) 는 덜 중요해져서 리지 회귀와 선형 회귀의 성능이 같아질 것이다.

# 3. Linear Classifier

```python
# Linear Regression & Linear Support Vector Machine
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10, 3))

for model, ax in zip([LinearSVC(max_iter=5000), LogisticRegression()], axes):
    clf = model.fit(X, y)
    
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5,
                                   ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')

axes[0].legend()
```

```python
# regularization argument : C
mglearn.plots.plot_linear_svc_regularization()
```

```python
# use breast cancer datset to see the performance of logistic regression on multiple dimension data
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    stratify=cancer.target,
                                                   random_state=42)
logreg = LogisticRegression(max_iter=5000).fit(X_train, y_train)

print(logreg.score(X_train, y_train))
print(logreg.score(X_test, y_test))
```

```python
# increase C(decrease the level of regularization) on logreg
logreg100 = LogisticRegression(C=100, max_iter=5000).fit(X_train, y_train)

print(logreg100.score(X_train, y_train))
print(logreg100.score(X_test, y_test))
```

```python
# decrease C(increase the level of regularization) on logreg
logreg001 = LogisticRegression(C=0.01, max_iter=5000).fit(X_train, y_train)

print(logreg001.score(X_train, y_train))
print(logreg001.score(X_test, y_test))
```

# 4. Multiclass Linear Classifier

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)

mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend(['class 0', 'class 1', 'class 2'])
```

```python
# learn linearsvc to classify multiclass dataset
linear_svc = LinearSVC().fit(X, y)

print(linear_svc.coef_.shape)
print(linear_svc.intercept_.shape)
```

```python
# decision boundary
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)

for coef, intercept, color in zip(linear_svc.coef_, linear_svc.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
    
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend(['class 0', 'class 1', 'class 2', 'class 0 boundary',
           'class 1 boundary', 'class 2 boundary'], loc=(1.01, 0.3))
```

```python
mglearn.plots.plot_2d_classification(linear_svc, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)

for coef, intercept, color in zip(linear_svc.coef_, linear_svc.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
    
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend(['class 0', 'class 1', 'class 2', 'class 0 boundary',
           'class 1 boundary', 'class 2 boundary'], loc=(1.01, 0.3))
```
